### Knobel Exercise 5.3
(Interacting Solitary Waves) Suppose $k_1$ and $k_2$ are positive numbers and set
$u_1 (x,t) = exp\{k^3_1 t - k_1 x\}$


$u_2 (x,t) = exp\{k^3_2 t - k_2 x\}$

$A = \frac{(k_1 - k_2)^2}{(k_1 + k_2)^2}$

let 
$ u = 12\frac{k^2_1 u_1 + k^2_2 u_2 + 2 (k_1 -k_2)^2 u_1 u_2 + A u_1 u_2(k^2_1 u_2 + k^2_1 u_1)}{(1 + u_1 + u_2 + A u_1 u_2)^2}$

Animate this function $u(x,t)$ for $ -10 < x < 10 $ and $-10 < t < 10$ to observe the behavior of this double soliton solution.

```python
import matplotlib.pyplot as plt
from matplotlib import animation
import numpy as np
import math
from mpl_toolkits.mplot3d import axes3d
```

```python
# define variables
k_1 = 1
k_2 = 2
# define functions

def u_1(x, t): return np.exp(((k_1) **3)*t-(k_1)*x)
def u_2(x, t): return np.exp((k_2**3)*t-(k_2)*x)
def A(k_1, k_2): return ((k_1 - k_2)**2) / ((k_1 + k_2)**2)
A = A(k_1,k_2)

def u(x,t): return 12*(((k_1)**2)*(np.exp(((k_1) **3)*t-(k_1)*x)) + (((k_2)**2)*(np.exp((k_2**3)*t-(k_2)*x))) + (2*(((k_1) - (k_2))**2)*(np.exp(((k_1) **3)*t-(k_1)*x)) * (np.exp((k_2**3)*t-(k_2)*x))   ) + (A * (np.exp(((k_1) **3)*t-(k_1)*x))*(np.exp((k_2**3)*t-(k_2)*x))*(((k_1)**2)*(np.exp((k_2**3)*t-(k_2)*x))) + (((k_2)**2)*(np.exp(((k_1) **3)*t-(k_1)*x))))) / ((1 + (np.exp(((k_1) **3)*t-(k_1)*x))+(np.exp((k_2**3)*t-(k_2)*x))+ (A * (np.exp(((k_1) **3)*t-(k_1)*x))*(np.exp((k_2**3)*t-(k_2)*x))))**2)


```
