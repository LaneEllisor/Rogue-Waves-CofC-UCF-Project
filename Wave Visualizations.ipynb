### Frame by Frame
This section produces 6 frames at times 0 to 5 for the function \
\
$u(x,t) = exp\{-(x-t)^2\}$ \
\
Here, it is shown at certain increments in time how the amplitude stays the same, but the wave's shape changes over position. 

```python
# 100 linearly spaced numbers
x = np.linspace(0,10,100)
t = np.linspace(0,10,100)
# the function, which is y = x^2 here
u_xt = np.exp(-(x-t)**2)

# setting the axes at the centre
fig, ax = plt.subplots(1, 6, figsize=(20,2))

#boundaries
ax[0].set_xlim([0,5])
ax[1].set_xlim([0,5])
ax[2].set_xlim([0,5])
ax[3].set_xlim([0,5])
ax[4].set_xlim([0,5])
ax[5].set_xlim([0,5])

# x labels
ax[0].set_xlabel('Position (x)')
ax[1].set_xlabel('Position (x)')
ax[2].set_xlabel('Position (x)')
ax[3].set_xlabel('Position (x)')
ax[4].set_xlabel('Position (x)')
ax[5].set_xlabel('Position (x)')

#y labels
ax[0].set_ylabel('u(x,t)')


# plot the function
ax[0].plot(x, np.exp(-(x)**2))
ax[1].plot(x, np.exp(-(x-1)**2))
ax[2].plot(x, np.exp(-(x-2)**2))
ax[3].plot(x, np.exp(-(x-3)**2))
ax[4].plot(x, np.exp(-(x-4)**2))
ax[5].plot(x, np.exp(-(x-5)**2))


# show the plot
plt.show()
```

### XT- Diagram
This diagram can be described as looking down on to the xt-plane, where the amplitude \
\
$u(x,t) = exp\{-(x-t)^2\}$\
\
is shown as the brighter blue colors. As time increases, the amplitude stays the same, but the position changes.

```python
# xt - plane view
import matplotlib.pyplot as plt
x = np.arange(-10, 10, 0.1)
t = np.arange(0, 10, 0.1)
xx, tt = np.meshgrid(x, t, sparse=True, )
u = np.exp(-(xx -tt)**2) 
h = plt.contourf(x,t,u, cmap = 'Blues_r') 
plt.xlabel('Position (x)')
plt.ylabel('Time (t)')


plt.show()
```

### Surface Plot
An amplitude function can be plotted as a 3D surface of $x$, $t$, and $u(x,t)$ in an xtu coordinate system. The colorbar on the side shows the color as a function of amplitude, with a lighter color being farther from the xt-axis. The function plotted below is \
\
$u(x,t) = exp\{-(x-t)^2\}$

```python
# Imports of tools
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

#definition of figure => make sure to specify 3d
fig = plt.figure(figsize = (10,5))
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.001)
t = np.arange(0, 5, 0.001)
X, t = np.meshgrid(X, t) #creating the mesh
U = np.exp(-(X-t)**2)


# Plot the surface.
surf = ax.plot_surface(X, t, U, cmap=cm.Blues_r,
                       linewidth=0, antialiased=False)

# Limit the z axis.
ax.set_zlim(0, 1.5)
#This is from the source code, but I don't know what it is yet lol
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.01f'))

# This is the creation of the color label by height
cbar = fig.colorbar(surf, shrink=0.5, aspect=5)
cbar.ax.set_ylabel('u(x,t)')
#labelling axes
ax.set_xlabel('Position (x)')
ax.set_ylabel('Time (t)')
ax.set_zlabel('u(x,t)')


plt.show()
```

### Surface Plot
This is the same plot as above, but with a different function,\
\
$u(x,t) = exp\{-(x-t)^2\}+exp\{-(x+t)^2\}$

```python
#definition of figure => make sure to specify 3d
fig = plt.figure(figsize = (10,5))
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-8,8, 0.005)
t = np.arange(0, 8, 0.005)
X, t = np.meshgrid(X, t) #creating the mesh
#define your wave function
U = np.exp(-(X-t)**2)+np.exp(-(X+t)**2)


# Plot the surface.
surf = ax.plot_surface(X, t, U, cmap=cm.Blues_r,
                       linewidth=0, antialiased=False)

# limit the z axis.
ax.set_zlim(0, 1.5)

#This is from the source code, but I don't know what it is yet lol
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.01f'))

# This is the creation of the color label by height
cbar = fig.colorbar(surf, shrink=0.5, aspect=5)
cbar.ax.set_ylabel('u(x,t)')

#labelling axes
ax.set_xlabel('Position (x)')
ax.set_ylabel('Time (t)')
ax.set_zlabel('u(x,t)')

plt.show()
```

```python
from matplotlib import animation

fig = plt.figure()
ax = plt.axes(xlim=(0,30), ylim=(0,3))
line, = ax.plot([],[], lw=2)
ax.set_xlabel('Position (x)')
ax.set_ylabel('u(x,t)')
def init():
    #create an empty plot/frame
    line.set_data([],[])
    return line,

xdata, ydata = [], []

def animate(t):
    #x and y values to plot
    x = np.linspace(0,30,500)
    u = np.exp(-(x-t)**2)
    
    xdata.append(x)
    ydata.append(u)
    
    line.set_data(x,u)
    return line,

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=500)
anim.save('func.gif')
```
