```python
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
```

```python
# Grid and test function
x,t = np.linspace(0,1, 10000), np.linspace(0,1, 10000)
X,T = np.meshgrid(x,t)
F = lambda X,T : 10*np.cos(np.pi *T) *np.sin(np.pi *X) + 3*np.cos(4 * np.pi *T) * np.sin(4*np.pi * X)
U_xt = F(X,T)

# 3D Surface plot
plt.figure(figsize = (20,20))
ax = plt.subplot(211, projection='3d')


# Normalise Y for calling in the cmap.
Ts = U_xt/U_xt.max()
cmap = plt.cm.Blues_r
ax.plot_surface(X, T, U_xt, facecolors=cmap(Ts))

#labels
ax.set_xlabel('Position (x)')
ax.set_ylabel('Time (t)')
ax.set_zlabel('u(x,t)')



# 2D Plot of slice of 3D plot 
# Normalise y for calling in the cmap.
ts = t/t.max()
plt.subplots(1,1)
plt.plot(x,U_xt[1,:], color=cmap(ts[1]))
plt.xlabel('Position (x)')
plt.ylabel('u(x,t)')
plt.show()
```

```python
from matplotlib import animation

fig = plt.figure()
ax = plt.axes(xlim=(0,1), ylim=(-13,13))
line, = ax.plot([],[], lw=2)
ax.set_xlabel('Position (x)')
ax.set_ylabel('u(x,t)')

def init():
    #create an empty plot/frame
    line.set_data([],[])
    return line,
plt.figure(figsize = (12,20))

xdata, ydata = [], []

def animate(t):
    #x and y values to plot
    x = np.linspace(0,1,5000)
    u = 10*np.cos(np.pi * t) *np.sin(np.pi *x) + 3*np.cos(4 * np.pi * t) * np.sin(4*np.pi * x)
    
    xdata.append(x)
    ydata.append(u)
    
    line.set_data(x,u)
    return line,
times = np.linspace(0, 11, 100)
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=times)
anim.save('K13_3a.gif')
```
