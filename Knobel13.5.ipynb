```python
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd

N = 7
x = np.linspace(0,1,100)

def f_n(x, n):
    f = np.sin((2*n - 1)*np.pi*x)
    coeff = (-1)**(n+1)/(2*n - 1)**2
    return coeff*f

def f_N(x, N):
    f = np.zeros_like(x)
    for i in range(1, N+1):
        f = f + f_n(x, i)
    return f
        
fig, ax = plt.subplots(1, 3, figsize= (20,2))

for i in range (0,3):
    ax[i].set_xlim([0,1])
    ax[i].set_xlabel(str(5*(2**i)) + 'terms')
    ax[i].plot(x, f_N(x, 5*(2**i)), 'r-')
    
plt.savefig('K13.5.png')
    
```
