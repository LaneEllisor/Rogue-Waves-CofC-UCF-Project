A sucsessful recreation of Matlab's "waterfall plot".

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x, t): return np.exp(-(x-t)**2)
```

### First, a standard wireframe plot
- Some small edits to shorten the code were made, specifically in the np.linspace call.
- Figure size and labels were created

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')


x = np.linspace(-10, 10, 1000)
t = np.linspace(0, 10, 500)
(X, T) = np.meshgrid(x, t)
Z = f(X, T)

plt.figure(figsize = (15,15))


#labels
ax.set_xlabel('Position (x)')
ax.set_ylabel('Time (t)')
ax.set_zlabel('u(x,t)')

# parameter "cstride=0" suppresses "wires" along columns of data
ignore_this_return_value = ax.plot_wireframe(X, T, Z)
```

### Next, an attempt at a waterfall plot

- Setting `cstride=0` in `plot_wireframe` suppresses "wires" in the second, "column" direction.
- More x values and fewer t values look best.
- Background was made transparent, but kept axes. Added more x values.

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection='3d')

# the last argument of t determines the number of time steps you want to see
x = np.linspace(-10, 10, 10000)
t = np.linspace(0, 10, 10)
(X, T) = np.meshgrid(x, t)
Z = f(X, T)



#labels
ax.set_xlabel('Position (x)')
ax.set_ylabel('Time (t)')
ax.set_zlabel('u(x,t)')

# make the panes transparent
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))

# make the grid lines transparent
ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)

#removes axes
#ax.set_axis_off()

ignore_this_return_value = ax.plot_wireframe(X, T, Z, cstride=0)
```

```python
def G(x, t): return np.exp(-(x-t)**2)+np.exp(-(x+t)**2)
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection='3d')

# the last argument of t determines the number of time steps you want to see
x = np.linspace(-30, 30, 10000)
t = np.linspace(0, 10, 10)
(X, T) = np.meshgrid(x, t)
Z = G(X, T)



#labels
ax.set_xlabel('Position (x)')
ax.set_ylabel('Time (t)')
ax.set_zlabel('u(x,t)')

# make the panes transparent
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))

# make the grid lines transparent
ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)

#removes axes
#ax.set_axis_off()

ignore_this_return_value = ax.plot_wireframe(X, T, Z, cstride=0)
```
